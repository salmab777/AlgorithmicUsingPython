## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
wat=pd.read_csv("../data/company_sales_data.csv")
profitlist = wat ['total_profit'].tolist()
monthlist = wat ['month_number'].tolist()
plt.plot(monthlist,profitlist,label = 'Last year profits per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000,500000])
plt.title('Company profit per month')
plt.show()
```

## Exercise 2

## Exercise 3

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
wat=pd.read_csv("../data/company_sales_data.csv")
monthnumber = wat['month_number'].tolist()
facecreamsales = wat['facecream'].tolist()
facewashsales = wat['facewash'].tolist()
toothpastesales = wat['toothpaste'].tolist()
bathingsoapsales = wat['bathingsoap'].tolist()
shampoosales = wat['shampoo'].tolist()
moisturizersales = wat['moisturizer'].tolist()

plt.plot(monthnumber,facecreamsales,label = 'Face cream sales data',marker='o',linewidth=3)
plt.plot(monthnumber,facewashsales,label = 'Face wash sales data',marker='o',linewidth=3)
plt.plot(monthnumber,toothpastesales,label = 'Tooth paste sales data',marker='o',linewidth=3)
plt.plot(monthnumber,bathingsoapsales,label = 'Bathing soap sales data',marker='o',linewidth=3)
plt.plot(monthnumber,shampoosales,label = 'Shampoo sales data',marker='o',linewidth=3)
plt.plot(monthnumber,moisturizersales,label = 'Moisturizer sales data',marker='o',linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(monthlist)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.title('Sales data')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
wat=pd.read_csv("../data/company_sales_data.csv")

labels = ['Toothpaste','Facewash','Facecream','Moisturizer','Shampoo','Bathing soap']
Salesdata = [wat['toothpaste'].sum(),wat['facewash'].sum(),wat['facecream'].sum(),wat['moisturizer'].sum(),wat['shampoo'].sum(), wat['bathingsoap'].sum()]
plt.pie(Salesdata,labels=labels,autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt  

wat = pd.read_csv("../data/company_sales_data.csv")
monthnumber= wat['month_number'].tolist()
bathingsoap= wat['bathingsoap'].tolist()
facewashsalesdata= wat['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthnumber, bathingsoap, label='Bathingsoap Sales Data',color='k',marker='o',linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(monthnumber, facewashsalesdata, label='Face Wash Sales Data',color='r',marker='o',linewidth=3)
axarr[1].set_title('Sales data of a facewash')

plt.xticks(monthnumber)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt  

wat=pd.read_csv("../data/company_sales_data.csv") 
monthnumber= wat['month_number'].tolist()

facecreamsalesdata = wat['facecream'].tolist()
facewashsalesdata = wat['facewash'].tolist()
toothpastesalesdata = wat['toothpaste'].tolist()
bathingsoapsalesdata = wat['bathingsoap'].tolist()
shampoosalesdata = wat['shampoo'].tolist()
moisturizersalesdata = wat['moisturizer'].tolist()

plt.plot([],[],color='m', label='Face cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(monthnumber,facecreamsalesdata,facewashsalesdata,toothpastesalesdata,bathingsoapsalesdata,shampoosalesdata, moisturizersalesdata, colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
